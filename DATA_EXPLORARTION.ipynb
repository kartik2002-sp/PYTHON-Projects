We will now handle the following data cleaning steps:

*   **Handle Datatype of Features**: Convert columns to their appropriate data types (e.g., numeric, datetime).

```python
df.isna().sum()             #   check for null values
```

```python
df[df.isna().any(axis=1)]                 # rows which are null
```

```python
df.dropna(axis = 0 , inplace = True)    # dropping nulls because they are less compared to dataset size
```

```python
df.head()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.reset_index(inplace = True)                # fixing the index numbering and dropping column 'index'
```

```python
df = df.drop(columns = 'index')
```

```python
df.head()
```

changing datatypes

```python
df = df[df['Order ID'] != 'Order ID']
df = df.astype({'Order ID':'int','Quantity Ordered':'int','Price Each':'float'})
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df.info()
```

```python
df.head(10)
```

**Data Exploration**
> : Extract relevant information from 'Order Date'

> : Create New Feature (Total sales)

> : Analyze Monthly Sales

> : Identify Top-Selling Products AND Worst Performing


### Separate time and date from order date

```python
df['Date'] = df['Order Date'].dt.date
df['Time'] = df['Order Date'].dt.time
df.drop(columns=['Order Date','Purchase Address'],inplace = True)
df.head()
```

### Creating Total Sales Column

```python
df['Total Sales'] = df['Quantity Ordered'] * df['Price Each']
df.head()
```

## Calculate Monthly Sales




```python
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month
df.head()
```

```python
monthly_sales = df.groupby('Month')['Total Sales'].sum().reset_index()
best_month = monthly_sales.idxmax()
max_sales = monthly_sales.max()
#print(f"The best month for sales was month {best_month} with total sales of ${max_sales:.2f}")
monthly_sales.head()
```

```python
fig = px.bar(monthly_sales, x='Month' , y='Total Sales')
fig.show()
```

## Summary of Monthly Sales Analysis

The analysis of monthly sales data revealed a clear trend in customer purchasing behavior throughout the year.

- **Best Performing Month**: December emerged as the best month for sales, achieving a total of **$4,613,443.34** in sales. This could be attributed to holiday shopping or end-of-year promotions.


### Identify Top-Selling Products

```python
top_selling_products =df.groupby('Product')['Total Sales'].sum().reset_index()
top_selling_products = top_selling_products.sort_values(by='Total Sales',ascending = False)
best_product = top_selling_products.iloc[0]['Product']
max_sales = top_selling_products.iloc[0]['Total Sales']
print(f"The best selling product is {best_product} with total sales of ${max_sales:.2f}")
top_selling_products.head(10)
```

### Worst Performing Products

```python
worst_product = top_selling_products.iloc[-1]['Product']
min_sales = top_selling_products.iloc[-1]['Total Sales']
print(f"The worst performing product is {worst_product} with total sales of ${min_sales:.2f}")
top_selling_products.tail(10)
```

### Total Units Sold

```python
product_units_ordered = df.groupby('Product')['Quantity Ordered'].sum().reset_index()
product_units_ordered = product_units_ordered.sort_values(by='Quantity Ordered', ascending=False)
print("Total units ordered for each product:")
display(product_units_ordered.head(10))
```

```python
fig1 = px.bar(product_units_ordered, y='Quantity Ordered', x='Product', text_auto='.2s',
            title="Total Units Ordered")
fig1.update_traces(textfont_size=12, textangle=0, textposition="outside", cliponaxis=False)
fig1.show()
```

### Plot total units ordered vs. total sales for each product

```python
merged_product_data = pd.merge(product_units_ordered, top_selling_products, on='Product')
merged_product_data.head(10)
```

```python
fig = go.Figure(data=[
    go.Bar(name='Quantity Ordered', x=merged_product_data['Product'], y=merged_product_data['Quantity Ordered']),
    go.Bar(name='Total Sales', x=merged_product_data['Product'], y=merged_product_data['Total Sales'])
])

fig.update_layout(barmode='group',
                  title='Quantity Ordered vs. Total Sales by Product',
                  xaxis_title='Product',
                  yaxis_title='Value')

fig.show()
```
