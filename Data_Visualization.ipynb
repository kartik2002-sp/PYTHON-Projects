```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
df = pd.read_csv('results_clean.csv')
df.head()
```

### handle missing values in your DataFrame df by dropping rows where specific columns are entirely null, and then reporting on the changes.

```python
initial_rows = df.shape[0]
df.dropna(subset=['year', 'type'], how='all', inplace=True)
dropped_rows = initial_rows - df.shape[0]
print(f"Dropped {dropped_rows} rows where both 'year' and 'type' were null.")
print(f"New DataFrame shape: {df.shape}")
print("\nNumber of missing values per column after dropping rows:")
```

###  performing data cleaning by imputing missing values with meaningful defaults

```python
df['team'].fillna('No Team', inplace=True)
df['medal'].fillna('No Medal', inplace = True)
df['place'].fillna(0, inplace = True)
```

```python
df.head()
```

```python
print("Number of missing values per column:")
print(df.isnull().sum())
```

```python
df['discipline'].fillna('Unknown', inplace=True)
df['noc'].fillna('Unknown', inplace=True)
print("Number of missing values per column after filling:")
print(df.isnull().sum())
```

### then converts some columns to more appropriate types

```python
print("Data types before conversion:")
print(df.dtypes)

# Convert 'year' and 'place' to integer types
df['year'] = df['year'].astype(int)
df['place'] = df['place'].astype(int)

# Convert object columns to category for memory optimization
for col in ['type', 'discipline', 'event', 'as', 'noc', 'team', 'medal']:
    if col in df.columns and df[col].dtype == 'object':
        df[col] = df[col].astype('category')

print("Data types after conversion:")
print(df.dtypes)
```

```python
df.head()
```

### **1. Top Disciplines by Unique Participants**

This section identifies and visualizes the Olympic disciplines that have attracted the most unique participants throughout the years. We calculate the number of unique `athlete_id` for each `discipline` and then display the top 10 disciplines in a bar chart to highlight which sports have seen the highest participation.

```python
discipline_participants = df.groupby('discipline')['athlete_id'].nunique().sort_values(ascending=False).reset_index()

print("Discipline with the most participants:")
print(discipline_participants.head(1))

print("Top 10 disciplines by number of participants:")
discipline_participants.head(10)
```

```python
# Get the top 10 disciplines for visualization
top_10_disciplines = discipline_participants.head(10).reset_index()

fig = px.bar(top_10_disciplines,
             x='discipline',
             y='athlete_id',
             title='Top 10 Disciplines by Number of Unique Participants',
             labels={'discipline': 'Discipline', 'athlete_id': 'Number of Unique Participants'},
             color='discipline')

fig.update_layout(xaxis={'categoryorder':'total descending'})
fig.show()
```

### **2. Year-wise Athlete Participation**

This analysis explores the trend of athlete participation over the years. We aggregate the number of unique athletes for each Olympic year and then visualize this trend using a line plot. This helps us understand how the scale of the Olympic Games, in terms of athlete involvement, has evolved over time.

```python
year_wise_participation = df.groupby('year')['athlete_id'].nunique().reset_index()

print("Year-wise athlete participation:")
print(year_wise_participation.head(10))
```

```python
fig = px.line(year_wise_participation,
              x='year',
              y='athlete_id',
              title='Year-wise Athlete Participation',
              labels={'year': 'Year', 'athlete_id': 'Number of  Athletes'})

fig.show()
```

### **3. Athlete Participation by Event Type**

Here, we examine the distribution of unique athlete participation across different types of Olympic events (e.g., Summer vs. Winter). We count the unique athletes for each event `type` and present this information in a bar chart to compare participation levels between event categories.

```python
type_wise_participation = df.groupby('type')['athlete_id'].nunique().reset_index()

print("Participation by event type:")
print(type_wise_participation)
```

```python
fig = px.bar(type_wise_participation,
             x='type',
             y='athlete_id',
             title='Athlete Participation by Event Type',
             labels={'type': 'Event Type', 'athlete_id': 'Number of Unique Athletes'},
             color='type')

fig.update_layout(xaxis={'categoryorder':'total descending'})
fig.show()
```

### **4. Top 10 National Olympic Committees (NOCs) by Total Medal Count**

This section focuses on the performance of different countries (represented by their NOCs) in terms of overall medal wins. We filter out entries where no medal was awarded, count the total medals for each NOC, and then display the top 10 performing NOCs using a bar chart.

```python
medal_counts = df[df['medal'] != 'No Medal'].groupby('noc')['medal'].count().sort_values(ascending=False).reset_index()

print("Top 10 NOCs by total medal count:")
print(medal_counts.head(10))
```

```python
import plotly.express as px

top_10_nocs = medal_counts.head(10)

fig = px.bar(top_10_nocs,
             x='noc',
             y='medal',
             title='Top 10 NOCs by Total Medal Count',
             labels={'noc': 'National Olympic Committee', 'medal': 'Total Medals'},
             color='noc')

fig.update_layout(xaxis={'categoryorder':'total descending'})
fig.show()
```

### **5. Top 10 NOCs by Medal Type Count (Gold, Silver, Bronze)**

Building on the previous analysis, this section provides a more detailed look into the medal distribution among the top NOCs. We break down the total medal count into Gold, Silver, and Bronze medals for each NOC and visualize this using a stacked bar chart. This allows for a granular understanding of which countries excel in specific medal categories.

```python
df['medal'].value_counts()
```

```python
medal_type_counts = df[df['medal'] != 'No Medal'].groupby(['noc', 'medal'], observed=False).size().unstack(fill_value=0)
medal_type_counts['Total'] = medal_type_counts[['Gold', 'Silver', 'Bronze']].sum(axis=1)
medal_type_counts = medal_type_counts.sort_values(by='Total', ascending=False)

print("Top 10 NOCs by medal type count:")
print(medal_type_counts.head(10))
```

```python
top_10_nocs_medals = medal_type_counts.head(10).reset_index()

fig = px.bar(top_10_nocs_medals,
             x='noc',
             y=['Gold', 'Silver', 'Bronze'],
             title='Top 10 NOCs by Medal Type Count',
             labels={'value': 'Number of Medals', 'noc': 'National Olympic Committee', 'variable': 'Medal Type'},
             color_discrete_map={'Gold': 'gold', 'Silver': 'silver', 'Bronze': '#CD7F32'})

fig.update_layout(xaxis={'categoryorder':'total descending'})
fig.show()
```
